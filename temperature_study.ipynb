```python
import google.generativeai as genai
import ipywidgets as widgets
from IPython.display import display, Markdown
from google.colab import userdata

# 1. Securely fetch and clean your API key
try:
    api_key = userdata.get('GOOGLE_API_KEY').strip()
    genai.configure(api_key=api_key)
except userdata.SecretNotFoundError:
    print("⚠️ ERROR: You forgot to add your API key!")
    print("Please go to the Secrets tab, add 'GOOGLE_API_KEY', and run this cell again.")
    raise SystemExit

# 2. Build the interactive UI elements
prompt_input = widgets.Textarea(
    value='Suggest a creative name for a pet dragon and explain why.',
    description='Prompt:',
    layout=widgets.Layout(width='80%', height='80px')
)

temp_slider = widgets.FloatSlider(
    value=1.0,
    min=0.0,
    max=2.0,
    step=0.1,
    description='Temperature:',
    layout=widgets.Layout(width='50%')
)

num_gens_slider = widgets.IntSlider(
    value=3,
    min=1,
    max=10,
    step=1,
    description='Responses:',
    layout=widgets.Layout(width='50%')
)

generate_btn = widgets.Button(
    description='Generate Responses',
    button_style='primary'
)

output_area = widgets.Output()

# 3. Define the logic to loop the API call
def generate_text(b):
    with output_area:
        output_area.clear_output()
        print(f"Generating {num_gens_slider.value} response(s). Please wait...\n")

        try:
            model = genai.GenerativeModel('gemini-3-flash-preview')

            # Loop based on the integer slider value
            for i in range(num_gens_slider.value):
                response = model.generate_content(
                    prompt_input.value,
                    generation_config=genai.GenerationConfig(
                        temperature=temp_slider.value
                    )
                )

                # Print a clear header for each response
                display(Markdown(f"### 🟢 Response {i + 1}"))
                display(Markdown(response.text))
                display(Markdown("--- \n")) # A visual separator line

        except Exception as e:
             output_area.clear_output()
             print(f"Error: {e}")

generate_btn.on_click(generate_text)

# 4. Display the UI
display(prompt_input, temp_slider, num_gens_slider, generate_btn, output_area)
```
